```python
import tensorflow as tf 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split

from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense 
from tensorflow.keras.layers import Dropout
from sklearn.metrics import confusion_matrix 
```

```python
data = pd.read_csv('Datasets/heart.csv') 
```

```python
data.head() 
```

```python
data.describe() 
```

```python
data.isnull().any() 
```

```python
# Assign Dependent and Independent variable
```

```python
X = data.iloc[:,:13].values 
y = data["target"].values 
```

```python
#Split data into Train and Test dataset
X_train,X_test,y_train, y_test = train_test_split(X,y,test_size = 0.3 , random_state = 0 ) 
```

```python
#Scale the data.
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler() 
X_train = sc.fit_transform(X_train) 
X_test = sc.transform(X_test) 
```

```python
# Building the Model
classifier = Sequential() 
classifier.add(Dense(activation = "relu", input_dim = 13, units = 8, kernel_initializer = "uniform")) 
classifier.add(Dropout(0.3))
classifier.add(Dense(activation = "relu", units = 14,kernel_initializer = "uniform")) 
classifier.add(Dropout(0.3))
classifier.add(Dense(activation = "relu", units = 14,kernel_initializer = "uniform"))
classifier.add(Dropout(0.3))
classifier.add(Dense(activation = "relu", units = 14,kernel_initializer = "uniform"))
classifier.add(Dropout(0.3))
classifier.add(Dense(activation = "sigmoid", units = 1, kernel_initializer = "uniform")) 
classifier.compile(optimizer = 'adam' , loss = 'binary_crossentropy', metrics = ['accuracy'] ) 
```

```python
#Fitting the Model
#classifier.fit(X_train , y_train , batch_size = 8 ,epochs = 100 ) 
```

```python
history = classifier.fit(X_train, y_train,validation_data = (X_test,y_test), epochs=100, batch_size=64)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Heart Disease prediction Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
#Performing prediction and rescaling
y_pred = classifier.predict(X_test) 
y_pred = (y_pred > 0.5) 
```

```python
#Confusion Matrix
cm = confusion_matrix(y_test,y_pred) 
cm 
```

```python
#Accuracy
accuracy = (cm[0][0]+cm[1][1])/(cm[0][1] + cm[1][0] +cm[0][0] +cm[1][1]) 
print(accuracy*100) 
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_test, y_pred)
```
