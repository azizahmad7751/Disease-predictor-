# Classification Models using sklearn 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
rf.score(X_train,y_train) 
```

```python
y_pred =rf.predict(X_test)
```

```python
y_pred
```

# Deeplearning using ANN

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
from keras.optimizers import Adam
```

```python
model = Sequential()
model.add(Dense(20,input_dim=30, kernel_initializer='he_uniform', activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(20,kernel_initializer='he_uniform',activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(20,kernel_initializer='he_uniform',activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(20,kernel_initializer='he_uniform',activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile( optimizer=Adam(lr=0.001), loss='binary_crossentropy',metrics = ['accuracy'])
 
    # fit the keras model on the dataset
history=model.fit(X, Y, batch_size=64, epochs=100)
```

```python
model.summary() 
```

```python
history = model.fit(X_train, y_train,validation_data = (X_test,y_test), epochs=100, batch_size=64)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
# Save the model as a pickle in a file 
joblib.dump(history, 'model1.pkl') 

# Load the model from the file 
model = joblib.load('model1.pkl')

```
